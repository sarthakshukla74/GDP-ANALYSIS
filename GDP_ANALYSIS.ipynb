**ANALYSING ARAB WORLD**

```python
df_ar=df[df['Country Name']=='Arab World']
```

```python
df_ar
```

```python
df_ar.plot(kind='line',x='Year',y='Value',figsize=(15,6),grid=True)
```

**ADDING GDP GROWTH COLOUMNS**

```python
growth=[0]
data=df_ar.values
for i in range(1,len(data)):
  prev=data[i-1][3]
  curr=data[i][3]
  growth.append(round((curr-prev)/prev*100,2))
```

```python
df_ar = df_ar.assign(growth=growth)
```

```python
df_ar
```

**GDP GROWTH OF EVERY COUNTRY**

```python
final=[]
for country_name in df['Country Name'].unique():
  df_country=df[df['Country Name']==country_name].copy()
  growth=[np.nan]
  data=df_country.values
  for i in range(1,len(data)):
    prev=data[i-1][3]
    curr=data[i][3]
    growth.append(round((curr-prev)/prev*100,2))
  df_country = df_country.assign(growth=growth)
  final.append(df_country)
```

```python
final
```

```python
df=pd.concat(final,axis=0)
```

```python
df.groupby('Country Name')['growth'].mean().plot(kind='bar',figsize=(15,6))
```

**GDP comparison between specific countries**

```python
def compare_gdp(lst):
  dfs=[]
  for i in lst:
    dfs.append(df[df['Country Code']==i])
    df_pr=pd.concat(dfs,axis=0)
  fig=px.line(df_pr,x='Year',y='Value',color='Country Name',title='GDP Comparison-'+'|'.join(lst))
  pyo.plot(fig,filename='|'.join(lst)+'.html')
```

```python
compare_gdp(['IND','USA'])
```
